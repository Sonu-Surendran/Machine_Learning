```python
X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])  #(m,n)
y_train = np.array([0, 0, 0, 1, 1, 1])
```

```python
def compute_logistic_cost(x_train, y_train, w, b):
  m, n = x_train.shape
  cost = 0

  for i in range(m):
    f_x = x_train[i]@w+b
    classify = 1/(1+np.exp(-f_x))
    classify = np.clip(classify, eps, 1 - eps)
    print(classify)
    cost += -y_train[i] * (np.log(classify)) - (1-y_train[i])* (np.log(1-classify))

  return cost/m
```

```python
w_tmp = np.array([1,1])
b_tmp = -3
eps = 1e-15
print(compute_logistic_cost(X_train, y_train, w_tmp, b_tmp))
```
