Learning Costing function

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300, 500])
weights = np.array([[100,100],[100,200],[200,100],[100,388]])

weights.shape[0]
```

Computing costing function

```python
def compute_cost(x,y,weight):
    m = x.shape[0]
    p = weight.shape[0]
    res_cost = None
    res_slope = []


    for j in range(p):
        w = weight[j][0]
        b = weight[j][1]
        cost = 0
        slope = []

        for i in range(m):
            fw_b = w*x[i]+b
            slope.append(fw_b)
            error = (fw_b - y[i])**2
            cost += error
        cost = cost/(2*m)

        if res_cost is None or res_cost>cost:
            res_cost = cost
            res_slope = slope.copy()
    return res_cost, res_slope
```

```python
cost, slope = compute_cost(x_train, y_train, weights)
print(cost)
print(slope)
```

```python
plt.scatter(x_train,y_train, marker="x", c="r", label="actuals")
plt.plot(x_train, slope, c="b", label="prediction")
plt.ylabel("price")
plt.xlabel("size")
plt.legend()
plt.show()
```
