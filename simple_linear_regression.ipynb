```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

This is training example for the linear regression model.

```python
x_train = np.array([4.0, 6.0])
y_train = np.array([200, 350])
```

```python
i = 0
x_i = x_train[i]
y_i = y_train[i]

print(f"input:{x_i}, target:{y_i}")
```

```python
plt.scatter(x_train, y_train, marker="p", c="r")
plt.title("housing prices")
plt.ylabel("Price in dollars unit 10000")
plt.xlabel("Size in sq(1000s):")
plt.show()
```

```python
w = 40
b = 30

def compute_model_output(x, w, b):
    f_wb = []
    for element in x:
        ans = w*element+b
        f_wb.append(ans)

    return f_wb
        
```

```python
res = compute_model_output(x_train, w, b)
```

```python
plt.plot(x_train, res, c="b", label="my prediction")
plt.scatter(x_train, y_train, marker="x", c="r", label="acutal prediction")
plt.ylabel("size in sqt (units in 1000)")
plt.xlabel("price in dollar (units in 1000")
plt.legend()
plt.show()
```
