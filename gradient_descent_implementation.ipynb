```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/housing.csv")
data.info()
```

```python
data.drop(["beds", "baths", "zip_code", "Unnamed: 0"], axis=1)
```

```python
plt.scatter(data["size"], data["price"])
plt.xlabel("size")
plt.ylabel("price")
plt.title("housing price")
plt.show()
```

```python
data["price"] = data["price"] // 10000
```

```python
def gradient_descent(data, lr, epcho):
    w = b = 0
    m = data.shape[0]
    x_train = data["size"]/1000
    y_train = data["price"]/1000

    plt.scatter(x_train, y_train, alpha=0.5)
    plt.show()

    for i in range(epcho):
        if i % 50 == 0:
            print(f"epcho{i}")
        per_weight = per_bias = 0
        for j in range(m):
            weight = (((w * x_train.iloc[j])+b) - y_train.iloc[j])*x_train.iloc[j]
            bias = (((w * x_train.iloc[j]) + b) - y_train.iloc[j])
            per_weight += weight
            per_bias += bias
        per_weight = per_weight/m
        per_bias = per_bias/m
        tmp_w = w - lr*per_weight
        tmp_b = b - lr*per_bias
        w, b = tmp_w, tmp_b

    return w,b
```

```python
w,b = gradient_descent(data, 0.001, 300)
x_train = data["size"]/1000
y_train = data["price"]/1000
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create the regression line
x_line = np.linspace(x_train.min(), x_train.max(), 100)
y_line = w * x_line + b

# Plot
plt.figure(figsize=(8, 5))
plt.scatter(x_train, y_train, color='blue', label='Training data', alpha=0.6)
plt.plot(x_line, y_line, color='red', linewidth=2, label=f'Fit: y = {w:.2f}x + {b:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fit')

plt.legend()
plt.grid(True)
plt.show()
```
